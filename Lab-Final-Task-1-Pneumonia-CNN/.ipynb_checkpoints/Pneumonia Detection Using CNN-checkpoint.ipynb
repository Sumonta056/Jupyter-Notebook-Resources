## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input
```

## Loading Datasets

```python
img_size = 150
batch_size = 30

train_datagen = ImageDataGenerator(rescale = 1/255.,brightness_range=[0.5,1.5], rotation_range=25,
                                  width_shift_range=0.1, height_shift_range=0.1 ,featurewise_center = True, featurewise_std_normalization=True, )
val_datagen = ImageDataGenerator(rescale = 1/255.)
test_datagen = ImageDataGenerator(rescale = 1/255.)


train_generator = train_datagen.flow_from_directory('./train',
                                                   target_size = (img_size, img_size),
                                                   batch_size = batch_size,
                                                   shuffle=True,
                                                    color_mode ="grayscale",
                                                   class_mode='binary')

val_generator = val_datagen.flow_from_directory('./val',
                                                   target_size = (img_size, img_size),
                                                   batch_size = batch_size,
                                                   shuffle=False,
                                                color_mode ="grayscale",
                                                   class_mode='binary')

test_generator = test_datagen.flow_from_directory('./test',
                                                   target_size = (img_size, img_size),
                                                   batch_size = batch_size,
                                                   shuffle=False,
                                                  color_mode ="grayscale",
                                                   class_mode='binary')
```

## Visualization of Training Data Samples

```python
labels = ['normal', 'pneumonia']
samples = train_generator.__next__()
images = samples[0]
target = samples[1]

plt.figure(figsize=(20,20))
for i in range(15):
    plt.subplot(5,5,i+1)
    plt.subplots_adjust(hspace=0.3, wspace=0.3)
    plt.imshow(images[i])
    plt.title(f"Class: {labels[int(target[i])]}")
    plt.axis('off')
```

## Creating CNN model

```python
model = Sequential()
model.add(Conv2D(16, (3,3), input_shape=(img_size,img_size, 1), activation='relu', padding='same'))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D())
model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

## Training the model

```python
history = model_fit = model.fit(train_generator, validation_data=val_generator, epochs=10)
```

## Plotting loss and accuracy curve

```python
plt.figure(figsize=(20,8))
plt.plot(model_fit.history['loss'])
plt.plot(model_fit.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

```python
test_classes = test_generator.classes
y_pred = model.predict(test_generator)
y_pred[y_pred>0.5] = 1
y_pred[y_pred<0.5] = 0
```

## Classification report

```python
print(f"Classification Report: \n {classification_report(test_classes, y_pred, target_names=labels)}\n")
print(f"Confusion Matric: \n {confusion_matrix(test_classes,y_pred)}\n\n")
print(f"Accuracy Score: \n {round((accuracy_score(test_classes, y_pred)),4)*100}%")
```

## Confusion Matrix

```python
plt.figure(figsize=(10,8))
sns.heatmap(confusion_matrix(test_classes,y_pred),annot=True,fmt='.3g',xticklabels=['NORMAL','PNEUMONIA'],
            yticklabels=['NORMAL','PNEUMONIA'],cmap='Blues')
plt.show()
```
